```python
import numpy as np
import matplotlib.pyplot as plt
#generate 1000 points from 0 to 2
t = np.linspace(0, 2, 1000)
frequency  = 2
amplitudes = [1.5, 2.5]


# i = 0, 1, 2..
# amp = 1.5, 2.5..
for i, amp in enumerate(amplitudes):
    signal = amp*np.sin(2*t*np.pi*frequency)
    plt.plot(t, signal, label = f'Signal {i+1} (Amplitude = {amp})', color = f'C{i}')
    

plt.axvline(x = 0, color = 'black')    
plt.axhline(y = 0, color = 'black')
plt.title("Lab 2 Assignment 1")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()
plt.show
```
